```python
import os
```

```python
directory = "./build/png/1024"
```

```python
with open('tabular.tex', 'w') as texfile:

    for dirPath, dirNames, fileNames in sorted(os.walk(directory)):
        print("Directory Path: ", dirPath)
        # section
        if os.path.relpath(dirPath, directory) == '.':
            pass
        else:
            
            print(os.path.relpath(dirPath, directory))
            title_raw = '\\section{{{0}}}\n'.format(os.path.relpath(dirPath, directory))
            title = title_raw.replace('_', ' ')
            texfile.write(title)
            
            fileNames = [n for n in fileNames]
            fileNames.sort()
            fileNames.remove('.DS_Store') if '.DS_Store' in fileNames else None

            texfile.write('\\begin{longtable}{m{120mm}m{40mm}} \n')
            for i in fileNames:
                print('\tFile Name: ', i)
                print('\t\tRoot:', os.path.join(dirPath, i))
                

                
                image = '\\includegraphics[width = 3cm]{{{0}}} \n'.format(os.path.join(dirPath, i))
                image_name = i.replace('_',' ')

                # Name, figure
                texfile.write('{0} & {1} \\\\ \n'.format(image_name, image))
                
            print('-' * 10)
            texfile.write('\\end{longtable} \n\n')

    
    texfile.close()
```
